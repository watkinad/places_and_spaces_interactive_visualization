```python
#Import necessay libraries
import pandas as pd
import json
from google.colab import files

# Load people.json (names and locations)
people_df = pd.read_json('people.json')

# Load submissions.xlsx (bios and creator images)
submissions_df = pd.read_excel('submissions.xlsx')

# Merge data together by keeping all people, add any matching bios/images
merged_df = pd.merge(people_df, submissions_df[['name', 'biography', 'creator image']],
                     on='name', how='left')

# Save to CSV
merged_df.to_csv('people and submissions.csv', index=False)

merged_df.head()

# Downloads to local device
#files.download('people and submissions.csv')
```
